```python
from textblob import TextBlob # Importo textblob
```

Realizo el análisis de sentimiento de cada tuit.

```python
for tuit in tuits:
    analysis = TextBlob(tuit)
    print(f"Tuit: {tuit}\nSentiment: {analysis.sentiment}\n")
```

Se ha imprimido el texto de cada tuit seguido de su análisis de sentimiento, que incluye la polaridad (rango de -1 a 1, donde valores negativos indican un sentimiento negativo, valores positivos indican un sentimiento positivo, y valores alrededor de 0 indican neutralidad) y la subjetividad (rango de 0 a 1, donde valores cercanos a 1 indican un texto muy subjetivo y valores cercanos a 0 un texto más objetivo).

Ahora quiero crear histogramas con las métricas obtenidas. Lo primero que hago es crear un DataFrame de Pandas basándome en la polaridad y subjetividad.

```python
import pandas as pd #importo pandas e introduzco los datos.

datos = {
    "Polaridad": [-0.14, 0.9, 0.14285714285714285, 0.2965833333333333, 0.0, -0.1722222222222222, 0.175, 0.6, -0.09999999999999999, 0.12727272727272726, -0.028124999999999997, -0.041666666666666664, 0.5, -0.18333333333333335, -0.32916666666666666, 0.0, 0.0, 0.04166666666666666, -0.08333333333333331],
    "Subjetividad": [0.56, 0.4, 0.26785714285714285, 0.5121666666666667, 0.0, 0.4222222222222222, 0.275, 0.3, 0.6499999999999999, 0.48636363636363633, 0.65, 0.75, 0.6, 0.3833333333333333, 0.7958333333333333, 0.06666666666666667, 0.0, 0.375, 0.8166666666666668]
}

df = pd.DataFrame(datos)
```

Creo un histograma para visualizar la distribución de polaridad de los tuits.

```python
import matplotlib.pyplot as plt #Importo matplotlib para crear los histogramas.

plt.figure(figsize=(10, 6))
plt.hist(df['Polaridad'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribución de la polaridad de los tuits')
plt.xlabel('Polaridad')
plt.ylabel('Número de tuits')
plt.axvline(df['Polaridad'].mean(), color='red', linestyle='dashed', linewidth=1)
plt.text(df['Polaridad'].mean(), 5, 'Media', color = 'red')
plt.show()
```

La línea roja discontinua indica la media de cada distribución, proporcionandoe un punto de referencia para entender la tendencia central de los datos.

De la misma manera, creo un histograma para la subjetividad.

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Subjetividad'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribución de la subjetividad de los tuits')
plt.xlabel('Subjetividad')
plt.ylabel('Número de tuits')
plt.axvline(df['Subjetividad'].mean(), color='red', linestyle='dashed', linewidth=1)
plt.text(df['Subjetividad'].mean(), 5, 'Media', color = 'red')
plt.show()
```
